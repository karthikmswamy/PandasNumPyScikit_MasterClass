## Library Imports

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.manifold import Isomap

import matplotlib.pyplot as plt
import numpy as np
```

## Scaling

```python
# Create example data
X_train = np.array([[1., -1., 2.],
                    [2., 0., 0.],
                    [0., 1., -1.]])

# Initialize and fit the scaler
scaler = StandardScaler().fit(X_train)

# Transform the data
X_scaled = scaler.transform(X_train)
print(X_scaled)
```

## Encoding

```python
# Create example data
dog_breeds = ['Beagle', 'Poodle', 'Husky', 'Labrador', 'Poodle']

# Initialize and fit the encoder
encoder = LabelEncoder().fit(dog_breeds)

# Transform the data
dog_breeds_encoded = encoder.transform(dog_breeds)

# Show the encoded data
print(dog_breeds_encoded)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create example data
dog_breeds = [['Beagle'], ['Poodle'], ['Husky'], ['Labrador'], ['Poodle']]

# Initialize and fit the encoder
encoder = OneHotEncoder().fit(dog_breeds)

# Transform the data
dog_breeds_encoded = encoder.transform(dog_breeds).toarray()

# Show the encoded data
print(dog_breeds_encoded)
```

## Imputing

```python
# Create example data with missing values
# Columns represent localities and rows represent dog breeds
X = np.array([[5, 3, np.nan],
              [2, 1, 4],
              [np.nan, 2, 3]])

# Initialize and fit the imputer with strategy 'mean'
imputer = SimpleImputer(strategy='mean').fit(X)

# Transform the data
X_imputed = imputer.transform(X)

# Show the imputed data
print(X_imputed)
```

## PCA

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Create an instance of PCA and specify the number of components
pca = PCA(n_components=2)

# Fit the model to your data
X_pca = pca.fit_transform(X)

# Plot the transformed data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```

## Isomap

```python
# Create an instance of Isomap and specify the number of components
isomap = Isomap(n_components=2)

# Fit the model to your data
X_isomap = isomap.fit_transform(X)

# Plot the transformed data
plt.scatter(X_isomap[:, 0], X_isomap[:, 1], c=y)
plt.xlabel('First Isomap Component')
plt.ylabel('Second Isomap Component')
plt.show()
```

## Gaussian Mixture Models

```python
from sklearn import datasets
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

iris = datasets.load_iris()
X = iris.data

# Create an instance of GaussianMixture and specify the number of components
gmm = GaussianMixture(n_components=3)

# Fit the model to your data
gmm.fit(X)

# Predict the labels for the data
labels = gmm.predict(X)

# Plot the data with the predicted labels
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```

## Clustering

```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

iris = datasets.load_iris()
X = iris.data

# Create an instance of KMeans and specify the number of clusters
kmeans = KMeans(n_clusters=3)

# Fit the model to your data
kmeans.fit(X)

# Predict the labels for the data
labels = kmeans.predict(X)

# Plot the data with the predicted labels
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```

## Outlier Detection

```python
from sklearn import datasets
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt

iris = datasets.load_iris()
X = iris.data

# Create an instance of LocalOutlierFactor and specify the number of neighbors
lof = LocalOutlierFactor(n_neighbors=20)

# Fit the model to your data
lof.fit(X)

# Predict the labels for the data
labels = lof.fit_predict(X)

# Plot the data with the predicted labels
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```

## Linear Models

```python
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# Load the California Housing dataset
california = fetch_california_housing()
X = california.data
y = california.target

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Standardize the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Create a Ridge Regression model
ridge = Ridge(alpha=1.0)

# Fit the model on the training data
ridge.fit(X_train, y_train)

# Make predictions on the test data
y_pred = ridge.predict(X_test)

# Evaluate the model using mean squared error
mse = mean_squared_error(y_test, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of the true values
plt.scatter(range(len(y_test[:100])), y_test[:100], c='blue', label='True Values', marker='.')

# Create a scatter plot of the predicted values
plt.scatter(range(len(y_pred[:100])), y_pred[:100], c='green', label='Predicted Values', marker='o')

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('True vs Predicted Values for Ridge Regression')
plt.legend()
plt.show()
```

## Support Vector Machines (SVM)

```python
from sklearn.svm import SVR

# Create an SVR model with a radial basis function (RBF) kernel
svr = SVR(kernel='rbf', C=1.0)

# Fit the model on the training data
svr.fit(X_train, y_train)

# Make predictions on the test data
y_pred = svr.predict(X_test)

# Evaluate the model using mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')
```

```python
# Create a scatter plot of the true values
plt.scatter(range(len(y_test[:100])), y_test[:100], c='blue', label='True Values', marker='.')

# Create a scatter plot of the predicted values
plt.scatter(range(len(y_pred[:100])), y_pred[:100], c='green', label='Predicted Values', marker='o')

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('True vs Predicted Values for SVMs')
plt.legend()
plt.show()
```

## Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsRegressor
# Create a KNeighborsRegressor model with 5 neighbors
knn = KNeighborsRegressor(n_neighbors=5)

# Fit the model on the training data
knn.fit(X_train, y_train)

# Make predictions on the test data
y_pred = knn.predict(X_test)

# Evaluate the model using mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')
```

```python
# Create a scatter plot of the true values
plt.scatter(range(len(y_test[:100])), y_test[:100], c='blue', label='True Values', marker='.')

# Create a scatter plot of the predicted values
plt.scatter(range(len(y_pred[:100])), y_pred[:100], c='green', label='Predicted Values', marker='o')

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('True vs Predicted Values for K-NNs')
plt.legend()
plt.show()
```

## Decision Trees

```python
from sklearn.tree import DecisionTreeRegressor

# Create a DecisionTreeRegressor model
tree = DecisionTreeRegressor()

# Fit the model on the training data
tree.fit(X_train, y_train)

# Make predictions on the test data
y_pred = tree.predict(X_test)

# Evaluate the model using mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')
```

## Hyperparameter Search

```python
from sklearn import svm, datasets
from sklearn.model_selection import GridSearchCV

# Load the iris dataset
iris = datasets.load_iris()

# Define the parameter grid
parameters = {'kernel':('linear', 'rbf'), 'C':[1, 10]}

# Create an instance of an SVM classifier
svc = svm.SVC()

# Create a GridSearchCV object with the SVM classifier and the parameter grid
clf = GridSearchCV(svc, parameters)

# Fit the GridSearchCV object on the iris data
clf.fit(iris.data, iris.target)
```

```python
# Print the best parameters found by the search
print(clf.best_params_)
```

## Scikit-learn Pipelines and Feature Unions

```python
from sklearn.datasets import fetch_california_housing
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR

# Load the California housing dataset
california_housing = fetch_california_housing(as_frame=True)
X = california_housing.data
y = california_housing.target

# Create a pipeline with two steps: StandardScaler and SVR
pipe = Pipeline([("scaler", StandardScaler()), ("svr", SVR())])

# Fit the pipeline on the California housing data
pipe.fit(X, y)
```

```python
# Create a feature union with two transformers: StandardScaler and PCA
union = FeatureUnion([("scaler", StandardScaler()), ("pca", PCA(n_components=2))])

# Create a pipeline with two steps: the feature union and an SVR regressor
pipe = Pipeline([("features", union), ("svr", SVR())])

# Fit the pipeline on the California housing data
pipe.fit(X, y)
```
